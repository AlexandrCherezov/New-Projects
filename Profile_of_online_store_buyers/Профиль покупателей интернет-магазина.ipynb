### Изменение данных в столбце `date` к формату дата-время (datetime).

#### Проверка наличия неявных дубликатов
Заказчик сообщил нам, что при выгрузке часть записей задвоилась (и даже затроилась) и ряд уже имеющихся заказов получили дубли с новым идентификатором пользователя.

Создадим небольшую табличку, которую назовем `doubles_orders` и поместим в нее данные об идентификатовах заказов и соответствующим им количестве идентификаторов покупателей.

````python
doubles_orders = data.groupby('order_id').agg({'customer_id':'nunique', 'quantity':'count'}
                                             ).sort_values('customer_id', ascending=False).reset_index()
doubles_orders = doubles_orders.query('customer_id > 1')
print("\033[34m""\033[1m""Количество задвоенных заказов:", doubles_orders.shape[0])
doubles_orders.columns = ['order_id', 'Колич.покупателей в заказе', "Колич.товаров в заказе"]
doubles_orders
````

````python
# функция для выведения списка идентификаторов покупателей, соответствующих одному и тому же заказу
def double_customers(line_number):
    # номер строки в таблице doubles_orders
    examp = [line_number]
    # пустой список для его заполнения идентификаторами покупателей
    list_n = []
    # переменная для управления номером строки при выборке идентификаторов покурателя
    index = 0
    # организуем цикл и отсеем строки со столбцами , содержащими идентификатор покупателя
    for ex in examp:
        print("\033[34m""Строка:", ex)
        t = doubles_orders.loc[ex, 'order_id']
        print("\033[34m""Идентификатор заказа:", t)
        print("\033[34m""Список идентификаторов покупалеля в дублированых строках (заказах):")
        v = data.loc[(data['order_id'] == t), 'customer_id'].reset_index()
    # организуем цикл для формирования списка с идентификаторами покупателя 
        for r in range(len(v)):
            list_n.append(v.loc[index,'customer_id'])
            index = index+1
            
        print(list_n)
        
    return (list_n)   
````

#### Гистограмма распределения частоты совершаемых событий
Построим гистограмму распределения частоты совершаемых событий
Создадим небольшую функцию для построения распределения по количеству тех или иных событий в разрезе групп

````python
def orders_distr(data_name, measure, ax):
    
    ax =  data_name.plot(y=measure
                    , kind = 'hist'
                    , figsize=(15,5)
                    ,  bins=20
                    , ax=ax
                    ,grid=True, )
    plt.title('Распределение по критерию\n{}\n  '.format(measure), fontsize=20, color = "b")
    plt.xlabel("Диапазон значений критерия", fontsize = 16, color = "b")
    plt.ylabel("Количество событий", fontsize = 16, color = "b")

    return
````

Совместим две диаграммы на одном графике,   
Вызовем функцию и построим графики распределения по событиям Количество товаров в заказе и цена товара

````python
# задаём размер сетки для графиков
plt.figure(figsize=(30, 5))
# в первой ячейке строим распределение по группе А
ax1 = plt.subplot(1, 2, 1)
# вызываем функцию 
orders_distr(data, 'quantity', ax1)
# во второй ячейке строим распределение по группе В
ax2 = plt.subplot(1, 2, 2)
# вызываем функцию
orders_distr(data, 'price', ax2)
````

Гистограммы пока показываю нам, что распределение количества товаров в заказе и стоимость товаров не является нормальным. Но, как мы говорили ранее - присутствуют выбросы.  
Чтобы выяснить их природу и определить границы выбросов обратимся в боксплоту.

#### Боксплот
Создадим еще одну функцию, на этот раз для построения "ящика с усами" 

````python
def orders_box(data_name, measure, ax, limit ):
    
    ax =  data_name.boxplot([measure], figsize=(5,5))
    
    plt.title('Распределение по критерию\n{}\n  '.format(measure), fontsize=20, color = "b")
    plt.ylabel("Значение критерия", fontsize = 16, color = "b")
    plt.ylim (0, limit)

    return
````

Совместим две диаграммы на одном графике,   
Вызовем функцию и построим графики бокс-плот

#### Точечная диаграмма числа событий
Чтобы определить верхнюю границу событий и отбросить выбросы, построим точечную диаграмму числа событий.

Для этой цели вновь создадим функци,  назовем ее ```scatter_views_and_purchase``` 

````python
def scatter_views_and_purchase(data_name, measure, ax, y_limit):
    x_values = pd.Series(range(0,len(data_name)))

    plt.scatter(x_values, data_name[measure])
    plt.title('Точечная диаграмма \nпо критерию\n{}\n \n'.format(measure), fontsize=20, color = "b")
    plt.xticks(rotation=0, fontsize=12, color = "b")
    plt.yticks(rotation=0, fontsize=12, color = "b")
    plt.xlabel("Расположение пользователя\n (номер) в датафрейме", fontsize=14, color = "b")
    plt.ylabel("Значение критерия", fontsize=14, color = "b")
    plt.ylim(0,y_limit)
    return
````

 #### Подготовим инструменты и данные
 Создадим две функции:
 - `event_in_time_plot` - для построения линейного графика. Для сглаживания шумов применим скользящее среднее
 - `event_in_time_bar`- для построения столбчатой диаграммы
 - `event_in_time_bar_year` - для построения столбчатой диаграммы с учетом в продажах года 

````python
# строим график, отражающий количество событий в зависимости от времени
# для сглаживания шумов применим скользящее среднее
def event_in_time_plot(data_name, eventname, time_interval):
    t1 = data_name.groupby(time_interval).agg({eventname:'nunique'}).sort_values(time_interval)
    
    ax1 = t1.rolling(14).mean().plot(grid=True
        , figsize=(25,10)
        ,  color='b')

    plt.title('Количество событий {} \nв зависимости от времени {}\n'.format(eventname, time_interval), fontsize=20, color = "b")
    plt.legend(fontsize=14, bbox_to_anchor=(0.5, 1))
    plt.xticks(rotation=0, fontsize=12, color = "b")
    plt.yticks(rotation=0, fontsize=12, color = "b")
    plt.xlabel('Дата', fontsize=14, color = "b")
    plt.ylabel(f"Количество событий {eventname}", fontsize=14, color = "b")
    #plt.show()
    return
````

````python
# строим столбчатую диаграмму, отражающую количество событий в зависимости от времени
def event_in_time_bar(data_name, eventname, time_interval, operation, ax):
    t1 = data_name.groupby(time_interval).agg({eventname:operation})#.sort_values(time_interval)

    ax = t1.plot(figsize=(25,20),
    kind ='bar',
    y = eventname,
    linewidth=5,
    alpha=0.7, 
    label='Операция при группировке: '+ operation,
    ax=ax
    )

    plt.title('Количество событий {} \nв зависимости от времени {}\n'.format(eventname, time_interval), fontsize=20, color = "b")
    plt.legend(fontsize=14, bbox_to_anchor=(0.5, 1))
    plt.xticks(rotation=0, fontsize=12, color = "b")
    plt.yticks(rotation=0, fontsize=12, color = "b")
    plt.xlabel('Дата', fontsize=14, color = "b")
    plt.ylabel(f"{operation} событий {eventname}", fontsize=14, color = "b")
    
    return
````

````python
# строим столбчатую диаграмму, отражающую количество событий в зависимости от времени
def event_in_time_bar_year(data_name, eventname, time_interval, operation):
    t1 = data_name.groupby(['year', time_interval]).agg({eventname:operation})#.sort_values(time_interval)

    ax = t1.plot(figsize=(25,10),
    kind ='bar',
    y = eventname,
    linewidth=5,
    alpha=0.7, 
    label='Операция при группировке: '+ operation
    )

    plt.title('Количество событий {} \nв зависимости от времени {}\n'.format(eventname, time_interval), fontsize=20, color = "b")
    plt.legend(fontsize=14, bbox_to_anchor=(0.5, 1))
    plt.xticks(rotation=0, fontsize=12, color = "b")
    plt.yticks(rotation=0, fontsize=12, color = "b")
    plt.xlabel('Дата', fontsize=14, color = "b")
    plt.ylabel(f"{operation} событий {eventname}", fontsize=14, color = "b")
    
    return
````

### Зависимость показателей от частоты совершения покупок

#### Зависимость среднего чека от частоты совершения покупок
Выясним зависимость среднего чека от частоты совершения покупок.  



 
Создадим переменную `cheque`, в которую сгруппируем заказы и для каждого заказа вычислим его размер (сумму заказа). Также в этой переменной подсчитаем количество товаров в заказе и укажеи идентификационный номер покупателя, совершившего данный заказ.

````python
cheque = data.groupby('order_id').agg({'revenue':'sum', 'product':'count', 'customer_id':'first'}).reset_index()
cheque.rename(columns = {'revenue':'сумма_заказа', 'product':'число_товаров_в_заказе'}, inplace = True )
cheque.sort_values("customer_id").head()
````

Вычислим средний чек для каждого покупателя за период времени, представленный в датафрейме.  
Сгруппируем пользователей в переменной `cheque`, рассчитаем среднее значение для столбца `сумма_заказа`, добавим колонку с количеством заказов. Сохраним результат в новую переменную `avg_cheque`.<br>
Также рассчитаем медианное (в данном случае не среднее, чтобы исключить слияние крупных по количеству товаров заказов) значение количества товаров в одном заказе

````python
avg_cheque = cheque.groupby('customer_id').agg({'order_id': 'count','сумма_заказа':'mean'
                                                , 'число_товаров_в_заказе':'median'}).reset_index()
avg_cheque.columns = ['customer_id', 'количество_заказов', 'средняя_сумма_заказа', 'медианное_число_товаров_в_заказе' ]
avg_cheque.sort_values('количество_заказов').sort_values('средняя_сумма_заказа', ascending=False).head()
````

В следующей строке кода обратим внимание, что:
- диапазон нормальных значений средней суммы заказа находится в пределах от 300 до 1347 единиц.
- максимальный размер заказа равен 49432 единиц.

Воспользуемся небольшой функцией `corr_score_sales` и построим диаграмму взаимной зависимости `hexbin` между суммой заказа покупателя и количеством заказов у покупателя

````python
def corr_score_sales(param_x, param_y, limit_x_floor, limit_x_ceiling, limit_y):
    
    avg_cheque.plot(x=param_x, y=param_y, kind='hexbin', gridsize=80, \
                                   sharex=False, grid=True, figsize=(10,3))
    plt.xlim(limit_x_floor, limit_x_ceiling)
    plt.ylim(0, limit_y)
    plt.title(f'Зависимость \nмежду {param_x} покупателя \nи {param_y}  у покупателя\n', fontsize=20, color = "b")
    plt.xticks(rotation=0, fontsize=12, color = "b")
    plt.yticks(rotation=0, fontsize=12, color = "b")
    plt.xlabel(param_x, fontsize=14, color = "b")
    plt.ylabel(param_y, fontsize=14, color = "b")
    plt.show()
    print("\033[34m"
        f'Козффициент корреляции между {param_x} покупателя \nи {param_y}  у покупателя =',\
      avg_cheque[param_x].corr(avg_cheque[param_y]).round(2)) 
    
    
    return
````

Произведем категоризацию данных в датафрейме `avg_cheque`:

- среднюю сумму заказа разобьем с шагом 1000 единиц для заказов менее иои равных 5000 единицам. Значения более 5000 отнесем в категорию "более 5000" Создадим столбец `сумма_заказа_катег`

````python
def categorize_purpose(cheque):
    
    for cheq in range(len(avg_cheque)):
        if cheque < 1000:
            return 'до 1000'
        if cheque < 2000:
            return 'до 2000'
        if cheque < 3000:
            return 'до 3000'
        if cheque < 4000:
            return 'до 4000'
        if cheque < 5000:
            return 'до 5000'
        else:
            return 'более 5000'
  
avg_cheque['сумма_заказа_катег'] = avg_cheque['средняя_сумма_заказа'].apply(categorize_purpose)
# добюавим столбец с общим размером выручки от покупателя
avg_cheque['выручка'] = avg_cheque['количество_заказов'] * avg_cheque['средняя_сумма_заказа']
avg_cheque.head()
````

````python
# функция для выведения списка идентификаторов покупателей, соответствующих одному и тому же заказу
def extract_goods(line_number, column):
    # номер строки в таблице doubles_orders
    examp = [line_number]
    # пустой список для его заполнения идентификаторами покупателей
    list_n = []
    # переменная для управления номером строки при выборке идентификаторов покурателя
    index = 0
    # организуем цикл и отсеем строки со столбцами , содержащими идентификатор покупателя
    for ex in examp:
        #print("\033[34m""Строка:", ex)
        #t = doubles_orders.loc[ex, 'order_id']
        #print("\033[34m""Идентификатор заказа:", t)
        #print("\033[34m""Список идентификаторов покупалеля в дублированых строках (заказах):")
        v = line_number
    # организуем цикл для формирования списка с идентификаторами покупателя 
        for r in range(len(v)):
            list_n.append(v.loc[index,column])
            index = index+1
            
        #print(list_n)
        
    return (list_n)   
````

## Сформулируем и проверим гипотезы

Сформулируем нулевую и альтернативную гипотезы.   
Проверим гипотезу с помощью статистического теста

### Гипотеза о различии суммы заказов покупателей, у которых есть в заказах товары из категорий переменной `category_name` и прочих покупателей.

Сформулируем нулевую гипотезу следующим образом: "Заказы, формируемые покупателями, у которых в заказах присутствовал хотя бы один раз и хотя бы один товар из  переменной `category_name`, имеют такую же сумму, как и заказы, формируемые прочими покупателями". Переводя в язык статистики: "Среднее значение первой выборки не имеет статистически значимого отличия от среднего значения второй выборки".

Сформулируем также и альтернативную гипотезу: "Заказы, формируемые покупателями, у которых в заказах присутствовал хотя бы один раз и хотя бы один товар из  переменной `category_name`, имеют сумму отличную от заказов, формируемых прочими покупателями". Альтернативная гипотеза говорит, что, скорее всего, нулевая гипотеза неверная, опровергает ее, но не доказывает обратное.

Выдвинутую гипотезу мы  проверим с помощью расчёта p-value — вероятности при верной нулевой гипотезе получить наблюдаемое значение или более удалённое от того, которое предположили в нулевой гипотезе.\
Если мы получим значение p-value меньше выбранного уровня статистической значимости, то нулевая гипотеза будет нами отвергнута в пользу альтернативной.\
Если же мы получим значение p-value больше уровня статистической значимости, то нулевая гипотеза нами не будет отвергнута.

Первым шагом проверим, распределены ли данные нормально - это даст нам возможность понять, какой из тестов следует использовать для проверки нашей гипотезы.

````python
def distribution(data_first, data_second, event_name, time_1, time_2):
    data_first.plot(y=event_name, kind='hist', bins=20, figsize=(8, 5), alpha = 0.5 )
    data_second.plot(y=event_name, kind='hist', bins=20, figsize=(8, 5), alpha = 0.4, grid=True)
    plt.title(f'Распределение по {event_name} \n по критериям {time_1} и {time_2} \n'
             , fontsize=20, color = "b")
    plt.xlabel(event_name, fontsize=14, color = "b")
    plt.ylabel('Количество заказов', fontsize=14, color = "b")
    plt.legend(labels=[f"Заказы {time_1}",f"Заказы {time_2}"],fontsize=14, bbox_to_anchor=(0.5, 1))
    plt.xticks(rotation=0, fontsize=12, color = "b")
    plt.yticks(rotation=0, fontsize=12, color = "b")
    plt.show()
    
    return
````

Используем стандартную библиотеку scipy.statsfrom.
Поскольку нам еще понадобится произвести этот тест повторно, для расчёта критерия Шапиро-Уилка создадим функцию test_shapiro

````python
def test_shapiro(sample):
    # зададим критический уровень статистической значимости
    alpha = 0.05  
    # применим метод Шапиро-Уилка
    results = st.shapiro(sample)
    # извлечем p_value - второе значение в массиве результатов (с индексом 1) - p-value
    p_value = results[1]  

    print('p-значение: ', p_value)

    if p_value < alpha:
        print('Отвергаем нулевую гипотезу: распределение не нормально')
    else:
        print('Не получилось отвергнуть нулевую гипотезу, всё нормально')
    return  
````

#### Посчитаем статистическую значимость различий методом Манна-Уитни

Для расчета создадим небольшую функцию test_mannwhitneyu.

````python
def test_mannwhitneyu (sample_1, sample_2):
    # уровень статистической значимости. Если p-value окажется меньше него, отвергнем гипотезу
    alpha = 0.05 
    # применим метод Манна-Уитни
    results = st.mannwhitneyu(sample_1, sample_2)

    print('p-значение:', "{0:.3f}".format(results[1]))

    if results.pvalue < alpha:
        print('Отвергаем нулевую гипотезу')
    else:
        print('Нет оснований отвергнуть нулевую гипотезу')
    return    
````

### Проверка гипотез.

1) гипотеза "Заказы, формируемые покупателями, у которых есть в заказах товары из переменной category_name, имеют такую же сумму, как и заказы, формируемые прочими покупателями" - не может быть нами подтверждена.   
2) гипотеза "Количество товаров в заказах, формируемых покупателями, у которых есть в заказах товары из переменной category_name, и  количество товаров в заказах, формируемых прочими покупателями"" - не может быть нами подтверждена.    
3) гипотеза "Заказы, совершающиеся в дневное время, имеют такую же сумму, как и заказы, совершающиеся в вечернее время" - не может быть нами подтверждена.   
4) гипотеза "в заказах, совершающихся в дневное время, число товаров в заказе такое же, как и в заказах, совершающихся в утреннее время" - не может быть нами отвергнута.

Суммируя результаты проверки гипотез мы делаем вывод, что:
- результаты проверок первой и второй ипотез подтверждают вывод, сделанный при исследовании сегментации покупателей о том, что для увеличения размера стоимости заказа рекомендуется предлагать покупателям на этапе формирования корзины предложение товаров из списка products_for_purposeподтверждается 
- меньшая выручка в вечернее время по сравнению с выручкой в дневное время может быть связана с размером суммы заказов
- отсутствуют статистически значимые различия в количестве товаров в заказе в утреннее и дневное время , стало быть, этот параметр не оказывает влияния на различие в выручке в это время суток
